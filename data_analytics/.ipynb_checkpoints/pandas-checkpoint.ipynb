```python
import faker
import random
fak = faker.Faker()
names = [fak.name() for i in range(10)]
emp_ids = ["a"+str(i) for i in range(10)]
salaries = [random.randint(2000,3000) for i in range(10)]
print(names)
print(emp_ids)
print(salaries)
```

```python
import faker
import random
fak = faker.Faker()
names = [fak.name() for i in range(10)]
emp_ids = ["a"+str(i) for i in range(10)]
salaries = [random.randint(2000,3000) for i in range(10)]
df = pd.DataFrame({"name":names,"emp_id":emp_ids,"salary":salaries})
print(df)
```

```python
import faker
import random
fak = faker.Faker()
data = []
for i in range(10):
    data.append({"name":fak.name(), "emp_id":"a"+str(i), "salary":random.randint(2000,3000)})
print(data)
df = pd.DataFrame(data)
print(df)
```
